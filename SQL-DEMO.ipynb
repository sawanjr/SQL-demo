## Connecting to the Database

```python
import mysql.connector

# Establish a connection to the database
conn = mysql.connector.connect(host="localhost" , user = "root" , password = "sawansqljr10" , database = "sawan_db")


# Create a cursor object
cursor = conn.cursor()
```

# Selecting

## Selecting whole data and printing Table

    -before adding the data to our database , lets see what we actually have in our database 

```python
# Select data from the Persons table
select_query = "SELECT * FROM Persons"
cursor.execute(select_query)
rows = cursor.fetchall() # all the data of table will store in the form of list in rows

# Process and print the selected data
for row in rows:
    person_id, last_name, first_name, address, city = row
    print(f"Person ID: {person_id}, Name: {first_name} {last_name}, Address: {address}, City: {city}")

# Close the cursor and connection
# cursor.close()
# conn.close()
```

we have three rows in our table `persons`

## Inserting Data

lets insert some more rows in our table `persons`

```python
# Insert data into the Persons table
insert_query = "INSERT INTO Persons (PersonID, LastName, FirstName, Address, City) VALUES (%s, %s, %s, %s, %s)"
data_to_insert = (5, 'starc', 'mitchell', 'australlia', 'melbourne')
cursor.execute(insert_query, data_to_insert)

# Commit the insert operation , means changes made here will be changed in our database
conn.commit() 
```

lets see our table again and check the changes made by us above

```python
# Select data from the Persons table
select_query = "SELECT * FROM Persons"
cursor.execute(select_query)
rows = cursor.fetchall() # all the data of table will store in the form of list in rows

# Process and print the selected data
for row in rows:
    person_id, last_name, first_name, address, city = row
    print(f"Person ID: {person_id}, Name: {first_name} {last_name}, Address: {address}, City: {city}")

# Close the cursor and connection
# cursor.close()
# conn.close()
```
